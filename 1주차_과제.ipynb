#1. 언어

##플리토 인공지능 번역 서비스

인공지능 번역기는 인간의 언어 능력을 모방하여 만들어진 기계이며 각 언어마다 단어, 문장 구조, 문법 등 충분한 언어 데이터와 
알고리즘을 토대로 인공지능을 반복적으로 학습시켜 만든다. 인공지능 번역기에 학습된 데이터 양이 많을수록 그리고 다양할수록, 번역의 품질은 높아지고 더욱 자연스러운 번역 결과를 제공할 수 있다.

사물 판별, 감정 분석, 태그 생성 등 알맞은 메타 데이터를 부여하는 ‘데이터 어노테이션’ 과정을 거쳐야 비로소 기계 학습형 데이터가 된다.인공지능 번역기가 똑똑하게 작동하기 위해서는 각 언어에 따른 형태의 다양성, 상황의 다양성, 언어 사용자의 다양성 등이 확보되어야 한다. 신조어, 구어, 일상생활에서의 표현 등 새롭고 다양하며 정확한 데이터는 인공지능 번역기가 ‘진짜’ 사람처럼 유행에 맞게 언어를 구사할 수 있도록 도와준다.

플리토의 인공지능 번역기는 ‘좋다’라는 한국어를 상황에 따라 영어로는 ‘good’, ‘nice’, ‘like’ 등으로 다르게 표현하고, ‘갑분싸(갑자기 분위기 싸해짐)’라는 신조어를 이해해 영어로 ‘the atmosphere suddenly became uncomfortable’라고 표현한다.



#2. 음성

##플리토 인공지능 스피커

무선 스피커의 일종으로서, 하나 이상의 인상적인 말(hot word)의 도움을 받아 상호작용 동작과 핸즈프리 활성화를 제공하는 가상 비서가 내장된 보이스 커맨드 디바이스이다

인공지능 스피커는 “음성 입력 및 인식 → 자연어 처리 → 인식 결과”의 단계를 거친다. 

1. 음성 입력 과정을 거쳐 사람의 음성을 컴퓨터가 이해할 수 있도록 텍스트화한다.
이 기술을 STT(Speech To Text)라고 하는데요. 어린아이가 받아쓰기를 하듯이 사람의 음성을 텍스트로 옮긴다. 주변 소음을 제외한 말소리를 파악하고 이어서 각 발음과 단어를 인식한 후 핵심어와 연결 단어를 인식해 입력해야한다.

2. 자연어 처리(Natural Language Processing, NLP) 과정을 거친다. 이 과정에서는 자연어에 대한 형태소 분석, 구문 해석, 의미 분석, 화용 분석 등을 통해 컴퓨터가 문장에 담긴 의도를 파악한다.

3. 기계는 인식 결과를 내놓는다. 인식된 요청에 따라 가장 최적의 결과를 찾아내고, TTS(Text to Speech) 기술을 통해 사람의 말소리처럼 응답한다.  친구에게 메시지를 보내라고 말했다면 전달한 내용으로 메시지를 전송하고, 날씨를 물어봤다면 인터넷에서 위치 기반의 날씨를 검색해 대답한다.

#3. 이미지

##수아랩(SUALAB) 수아키트(SuaKit)

인공지능 이미지 해석 기술을 기반으로 기존의 머신비전 기술을 향상시킨 딥러닝 머신비전 검사 솔루션을 제공한다. 
정상 제품의 사진과 불량품 사진을 AI가 스스로 학습해 불량품을 쉽게 판별해낼 수 있다. 사진만으로도 학습 데이터를 쉽게 입력할 수 있어 효율적인 제품이다. 육안 검사 대신 AI를 활용하는 데도 5픽셀(1.3㎜) 이하의 작은 불량까지 걸러낼 수 있다.

###실사용 분야
* 전기·전자: 반도체, PCB, 디스플레이, 태양광 셀, 배터리, 카메라 모듈 등
* 자동차: 자동차 부품(베어링, 실린더 등), 산업용 섬유, 도장 검사 등
* 식품; 식음료 이물 검사, 패키징 외관 검사, 식자재 불량 검사 등
* 화학: 2차 전지, 산업용 필름 등
* 기타: 철강, 비파괴 검사(X-ray), 건축물 외관 검사, 기차 부품 검사 등

#4. 자율주행 

##테슬라 오토파일럿

Tesla가 제공하는 시스템은 운전석에 있는 사람이 아무런 행동을 취하지 않아도 단거리 및 장거리 주행을 수행할 수 있도록 고안되었다.
차에 타서 목적지를 말하기만 하면 되고, 아무 말도 하지 않으면 자동차는 달력을 확인해서 가정한 목적지로 가거나, 달력에 일정이 따로 없으면 집으로 간다.
Tesla는 최적의 경로를 계산해서 도시 거리 (차선 표시가 없어도)를 주행하며, 신호등, 정지 신호 및 로터리가 있는 복잡한 교차로도 주행할 수 있고, 많은 차량들이 빠르게 이동하는 밀집되고 혼잡한 고속도로에서도 주행할 수 있다.
목적지에 도착해서 입구에 내리면 차량이 주차 탐색 모드로 진입한 후, 자동으로 주차 공간을 찾아서 주차한다. 핸드폰을 탭해서 다시 차량을 부를 수 있다.
